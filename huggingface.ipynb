```python
# pipeline
# sentiment analysis
from transformers import pipeline
classifier=pipeline('sentiment-analysis')
subject=classifier('i am fine')
```

```python
subject
```

```python
classifier(['i hate you' , 'i need you'])
```

```python
# zero shot classifier
classifier=pipeline('zero-shot-classification')
classifier('i love you',candidate_labels=['people','education','affection'])
```

```python
# text generation
generator=pipeline('text-generation')
generator('i love you until my ')
```

```python
generator=pipeline('text-generation',model='distilgpt2')
generator('i love you until my ')
```

```python
#  NER
nr=pipeline('ner',grouped_entities=True)
nr('i am sachin from karur')
```

```python
# Question Answer

qa=pipeline('question-answering')
qa(
    question='where i work',
    context='i am sachin from karur and i work in guvi'
)
```

```python
# preprocessing the text
from transformers import AutoTokenizer
model_checkpoint='distilbert-base-uncased-finetuned-sst-2-english'
token=AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
raw_input=['i willing to attend class','today is holiday']
result=token(raw_input,padding=True,truncation=True,return_tensors='pt')#pt=pytorch
```

```python
print(result)
```

```python
from transformers import AutoModel
model_checkpoint='distilbert-base-uncased-finetuned-sst-2-english'
model=AutoModel.from_pretrained(model_checkpoint)
```

```python
import pandas as pd
output=model(**result)
print(output.last_hidden_state.shape)
```

```python
from transformers import AutoModelForSequenceClassification
model_checkpoint='distilbert-base-uncased-finetuned-sst-2-english'
model=AutoModelForSequenceClassification.from_pretrained(model_checkpoint)
output=model(**result)
```

```python
print(output.logits.shape)
```

```python
import torch

prediction=torch.nn.functional.softmax(output.logits,dim=-1)
```
