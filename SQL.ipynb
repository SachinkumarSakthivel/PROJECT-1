```python
mycursor.execute("use hospital")
#mycursor.execute("select age, pregnancies ,count(*) over (partition by pregnancies order by age desc) from diab_1")
mycursor.execute("select  distinct(pregnancies) , avg(age) over (partition by pregnancies order by age desc) as countw from diab_1")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[x[0] for x in mycursor.description], tablefmt="psql"))
                 
```

```python
mycursor.execute("select age , pregnancies, rank() over (partition by pregnancies order by age desc) as rank from diab_1")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt="psql"))
```

```python
mycursor.execute("select age , pregnancies, dense_rank() over (partition by pregnancies order by age desc) as dense_rank from diab_1")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt="psql"))
```

```python
mycursor.execute("select pregnancies,outcome ,count(*) over (partition by pregnancies  order by outcome) as countw from diab_1 group by pregnancies,outcome ")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt="psql"))
```
